# 1) Write a python program to display all the header tags from wikipedia.org.

```python
# Import necesary library

from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
def header(url):
    header=[]
    r= BeautifulSoup(requests.get(url).content).find_all(['h1','h2','h3','h4','h5','h6'])
    for i in r:
        header.append(i.text.replace('\n',""))
    return pd.DataFrame(header,columns=['Header Tags'])
    
```

```python
header('https://en.wikipedia.org/wiki/Main_Page')
```

# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
def header(url):
    movie=[]
    rating=[]
    year=[]
    soup= BeautifulSoup(requests.get(url).content)
    
    for i in soup.find_all('div',class_="lister-item-content"):
        movie.append(i.find('a').text)
    
    for i in soup.find_all('div',class_="ipl-rating-star small"):
        rating.append(i.text.strip())
        
    for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
        year.append(i.text[-5:-1])
    
    data= pd.DataFrame({"Movie Name":movie, "Ratings":rating,"Year of Release":year})
    data.index+=1
    return data
```

# 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp

```python
def product(url):
    name=[]
    price=[]
    discount=[]
    soup= BeautifulSoup(requests.get(url).content)
    
    for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
        name.append(i.text.strip())
    
    for i in soup.find_all('div',class_="Card__BaseCard-sc-b3n78k-0 iLPHgK NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr"):
        price.append(i.find('h5').text.replace('₹','' ))
        
    for i in soup.find_all('div',class_="Card__BaseCard-sc-b3n78k-0 iLPHgK NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr"):
        discount.append(i.find('span').text)
    
    data= pd.DataFrame({"Product Name":name, "Price":price,"Discount in %":discount})
    return data
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.

```python
def ODI_team(url):
    
    soup= BeautifulSoup(requests.get(url).content)  # Get website content
    team=[]
    matches=[]
    points=[]
    ratings=[]
# Take 1st Rank,Matches,Points,Ratings from Banner(table content)

    for i in soup.find_all('span',class_="u-hide-phablet"):
        team.append(i.text)
    
    for i in soup.find_all('td',class_="rankings-block__banner--matches"):
        matches.append(i.text)
    
    
    for i in soup.find_all('td',class_="rankings-block__banner--points"):
        points.append(i.text)
    
    for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
        ratings.append(i.text.strip())
    
# Take Matches, Points, Ratings (2-10)
    table=[]
    for i in soup.find_all('tr',class_="table-body"):
        table.append(i.text.split())

    for i in table:
        matches.append(i[-3])
        points.append(i[-2])
        ratings.append(i[-1])
    
    data=pd.DataFrame({'Team':team[:10],'Matches':matches[:10],'Points':points[:10],'Ratings':ratings[:10]})
    data.index +=1
    return data



def ODI(url):
    
    soup= BeautifulSoup(requests.get(url).content)
    player=[]
    team=[]
    ratings=[]

# take 1st rank from Banner(table contect)
    for i in soup.find_all('div',class_="rankings-block__banner--name-large"):
        player.append(i.text)
    
    for i in soup.find_all('div',class_="rankings-block__banner--nationality"):
        team.append(i.text.strip())
    
    
    for i in soup.find_all('div',class_="rankings-block__banner--rating"):
        ratings.append(i.text)
    
        
# take remaining content
    for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
        player.append(i.text.strip())

    for i in soup.find_all('span',class_="table-body__logo-text"):
        team.append(i.text.strip())   
    
    for i in soup.find_all('td',class_="table-body__cell rating"):
        ratings.append(i.text.strip())  
        
    data= pd.DataFrame({"Player":player[:10],"Team":team[:10],'Ratings' :ratings[:10] })
    data.index +=1
    return data
```

# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
def coreyms_details(url):
    soup= BeautifulSoup(requests.get(url).content)  # Get website content

    data= requests.get('https://coreyms.com/')
    soup= BeautifulSoup(data.content)
    heading=[]
    date=[]
    content=[]
    video=[]

    for i in soup.find_all('h2',class_="entry-title"):
        heading.append(i.text)
        
    for i in soup.find_all('time',class_="entry-time"):
        date.append(i.text)
        
    for i in soup.find_all('div',class_="entry-content"):
        content.append(i.find('p').text)
        
    for i in soup.find_all('span', class_="embed-youtube"):
        video.append(i.find('iframe')['src'].strip())
        
    video.insert(4,'NA')
        
    print("Length of All columns\nHeading= ",len(heading),'\nDate= ',len(date),'\nContent Length =',len(content),'\nVideo Length= ',len(video)  )  
        
    data= pd.DataFrame({'Heading':heading,'Date': date, "Content": content,"Video Link": video  })  
    data.index +=1
    return data
    
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
# Import necesary library

from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
def scrapping(url):
    #create enmpt list to store variables
    restro=[]
    cuisine=[]
    location=[]
    ratings=[]
    image=[]
    
    soup=BeautifulSoup(requests.get(url).content)
    
    for i in soup.find_all('div', class_="restnt-info cursor"):
        restro.append(i.text)
    
    for i in soup.find_all('span',class_="double-line-ellipsis" ):
        cuisine.append(i.text.split('|')[1])
    
    for i in soup.find_all('div',class_="restnt-loc ellipsis"  ):
        location.append(i.text)
    
    for i in soup.find_all('div',class_="restnt-rating rating-4"):
        ratings.append(i.text)
    
    for i in soup.find_all('img',class_="no-img" ):
        image.append(i['data-src'])
    
    data= pd.DataFrame({"Restaurant name":restro, "Cuisine":cuisine,"Location":location,"Ratings":ratings,"Image Url":image  })
    data.index+=1
    
    return data
```

```python
scrapping( 'https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
def product(url):
    data= requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
    soup= BeautifulSoup(data.content)
    product=[]
    price=[]
    image=[]
    for i in soup.find_all('div',class_="productCardBox"):
        product.append(i.find('h3').text)
        
    for i in soup.find_all('div',class_="productPriceBox clearfix"):
        price.append(i.find('span').text.replace("₹",""))
        
    for i in soup.find_all('div',class_="productCardImg false"):
        image.append(i.find('img')['src'])
        
        
    data= pd.DataFrame({'Product Name':product,'Price': price, "Image URL": image  })  
    data.index +=1
    
    return data

    
    
```
